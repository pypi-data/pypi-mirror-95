```python
algos = all_data.algo.unique()
algos = algos.astype('str')
algos
```

```python
stats_by_algo = {}
for algo in algos:
    one_algo_data = all_data[all_data['algo'] == algo]
    stats_by_algo[algo] = one_algo_data.groupby('algo')['time'].describe()
```

```python
data_algo_rep = all_data.groupby(['algo', 'replicon'])
```

```python
stats = {}
for key, df in data_algo_rep:
    algo, rep = key
    time = df.time.mean()
    print(f"{algo} {rep} {time:.2f}")
    print("========================")
```

```python
stats_mean = pd.DataFrame(columns=['replicon', 'algo', 'time'])
for key, df in data_algo_rep:
    algo, rep = key
    time = df.time.mean()
    stats_mean = stats_mean.append({'algo':algo, 'replicon':rep, 'time': time}, ignore_index=True)
stats_mean
```

```python
bnb = stats_mean[stats_mean.algo == 'bnb']
bnb_max_bound = stats_mean[stats_mean.algo == 'bnb_max_bound']
bnb_max_bound_all_solution = stats_mean[stats_mean.algo == 'bnb_max_bound_all_solution']
bnb_iter = stats_mean[stats_mean.algo == 'bnb_iter']
mip = stats_mean[stats_mean.algo == 'mip']
nx = stats_mean[stats_mean.algo == 'nx']
```

```python
sys_size = pd.DataFrame([['GCF_000005845', 30],
                         ['GCF_000006725', 153],
                         ['GCF_000006745', 15],
                         ['GCF_000006765', 64],
                         ['GCF_000006845', 32],
                         ['GCF_000006905', 2],
                         ['GCF_000006925', 1],
                         ['GCF_000006945', 12],
                        ], columns=['replicon', 'system'])
sys_size = sys_size.sort_values(by='system')
sys_size
```

```python
bnb
```

```python

bnb = bnb.set_index('replicon')
bnb = bnb.reindex(index=sys_size['replicon'])

bnb_max_bound = bnb_max_bound.set_index('replicon')
bnb_max_bound = bnb_max_bound.reindex(index=sys_size['replicon'])

bnb_max_bound_all_solution = bnb_max_bound_all_solution.set_index('replicon')
bnb_max_bound_all_solution = bnb_max_bound_all_solution.reindex(index=sys_size['replicon'])

bnb_iter = bnb_iter.set_index('replicon')
bnb_iter = bnb_iter.reindex(index=sys_size['replicon'])

mip = mip.set_index('replicon')
mip = mip.reindex(index=sys_size['replicon'])

nx = nx.set_index('replicon')
nx = nx.reindex(index=sys_size['replicon'])
```

```python
bnb.reindex(index=sys_size['replicon'])
```

```python
fig = plt.figure(figsize=(10, 8))
plt.plot(sys_size.system, bnb.time, "o", linestyle="-", color="blue", label="bnb")
plt.plot(sys_size.system, bnb_max_bound.time, "v", linestyle="-", color="green", label="bnb_max_bound")
plt.plot(sys_size.system, bnb_max_bound_all_solution.time, "s", linestyle="-", color="darkgoldenrod", label="bnb_max_bound_all_solution")
plt.plot(sys_size.system, bnb_iter.time, "+", linestyle="-", color="magenta", label="bnb_iter")
plt.plot(sys_size.system, mip.time, "D", linestyle="-", color="mediumaquamarine", label="mip")
plt.plot(sys_size.system, nx.time, "*", linestyle="-", color="red", label="nx")
plt.legend(fontsize=12)
plt.title("algos comparison", fontsize=20)
plt.xlabel("systems nb", fontsize=15)
plt.ylabel("Execution time (in sec)", fontsize=15)
plt.grid()
```

```python
fig.savefig('algo_exec_time_1.png', bbox_inches='tight')
```
